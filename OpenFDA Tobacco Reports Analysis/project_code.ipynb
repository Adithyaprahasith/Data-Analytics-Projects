```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score


# Define simple functions to create target variables
def is_respiratory(text):
    text = str(text).lower()
    respiratory_patterns = ['breath', 'cough', 'lung', 'wheez', 'asthma','throat','dyspnea','respiratory','pulmonary','Nasal','Pneumonia']
    return 1 if any(pattern in text for pattern in respiratory_patterns) else 0

def is_cardiovascular(text):
    text = str(text).lower()
    cardiovascular_patterns = ['heart', 'cardiac', 'chest','stroke','blood pressure','pulse']
    return 1 if any(pattern in text for pattern in cardiovascular_patterns) else 0

def is_neurological(text):
    text = str(text).lower()
    neuro_patterns = ['seizure', 'headache', 'dizz', 'brain','migraine','Unconsciousness','anger','confusion','fatigue']
    return 1 if any(pattern in text for pattern in neuro_patterns) else 0

def classify_disease(text):
    text = str(text).lower()

    if any(p in text for p in ['breath', 'cough', 'lung', 'wheez', 'asthma','throat','dyspnea','respiratory','pulmonary','Nasal','Pneumonia']):
        return 'Respiratory'
    elif any(p in text for p in ['seizure', 'headache', 'dizz', 'brain','migraine','Unconsciousness','anger','confusion','fatigue']):
        return 'Neurological'
    elif any(p in text for p in ['heart', 'cardiac', 'chest','stroke','blood pressure','pulse']):
        return 'Cardiovascular'
    elif any(p in text for p in ['nausea', 'vomit', 'stomach','bladder','abdominal']):
        return 'Gastrointestinal'
    elif any(p in text for p in ['fever', 'flu', 'sick','weak']):
        return 'General'
    elif any(p in text for p in ['tooth', 'burn', 'skin','teeth','tongue','mouth']):
        return 'skin or mouth burns'
    else:
        return 'Other'

# Create target variables
df['is_respiratory'] = df['reported_health_problems'].apply(is_respiratory)
df['is_cardiovascular'] = df['reported_health_problems'].apply(is_cardiovascular)
df['is_neurological'] = df['reported_health_problems'].apply(is_neurological)
df['disease_class'] = df['reported_health_problems'].apply(classify_disease)
```

```python
def classify_product_problems(text):
    text = str(text).lower()

    if any(p in text for p in ['appearance', 'foreign','taste','smell','label','wrong']):
        return 'foreign products or continamination issues'
    elif any(p in text for p in ['explode','caught','fire','burn','overheat']):
        return 'safety issues'
    elif any(p in text for p in ['damage', 'defective', 'leak','broke']):
        return 'product quality issues'
    else:
        return 'Other'
df['products_problems_class'] = df['reported_product_problems'].apply(classify_product_problems)
```

```python
def classify_products(text):
    text = str(text).lower()
    if any(p in text for p in ['electronic','vape']):
        return 'e-cigarettes or vaping'
    elif any(p in text for p in ['cigar','cigarette',]):
        return 'cigars or cigarettes'
    elif any(p in text for p in ['chewing','snuff','snus','dissolvables','mixtures','gutka','pipe']):
        return 'oral tobacco'
    else:
        return 'other tobacco products'
df['products_class'] = df['tobacco_products'].apply(classify_products)
```

```python
# Print class distribution
print("Binary class distribution:")
print(df['is_respiratory'].value_counts())
print("\nMulti-class distribution:")
print(df['disease_class'].value_counts())

# Combine text features
df['text_features'] = df['reported_health_problems'] + ' ' + df['reported_product_problems']

# Split data for binary classification
X = df['text_features']
y_binary = df['is_respiratory']
y_multi = df['disease_class']

X_train, X_test, y_train_binary, y_test_binary = train_test_split(
    X, y_binary, test_size=0.3
)

X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
    X, y_multi, test_size=0.3
)

# Create and train binary classification model
binary_pipeline = Pipeline([
    ('vectorizer', CountVectorizer(max_features=1000)),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

print("Training binary classifier...")
binary_pipeline.fit(X_train, y_train_binary)
y_pred_binary = binary_pipeline.predict(X_test)

# Evaluate binary model
binary_accuracy = accuracy_score(y_test_binary, y_pred_binary)
print(f"\nBinary Classification Results:")
print(f"Accuracy: {binary_accuracy:.4f}")
print(classification_report(y_test_binary, y_pred_binary))

# Add confusion matrix for binary classification
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Create binary confusion matrix
binary_cm = confusion_matrix(y_test_binary, y_pred_binary)

# Plot binary confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(binary_cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Respiratory', 'Respiratory'],
            yticklabels=['Non-Respiratory', 'Respiratory'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Binary Classification (Respiratory vs Non-Respiratory)')
plt.tight_layout()
plt.show()

# Create and train multi-class model
multi_pipeline = Pipeline([
    ('vectorizer', CountVectorizer(max_features=1000)),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

print("\nTraining multi-class classifier...")
multi_pipeline.fit(X_train_multi, y_train_multi)
y_pred_multi = multi_pipeline.predict(X_test_multi)

# Evaluate multi-class model
multi_accuracy = accuracy_score(y_test_multi, y_pred_multi)
print(f"\nMulti-class Classification Results:")
print(f"Accuracy: {multi_accuracy:.4f}")
print(classification_report(y_test_multi, y_pred_multi))

# Add confusion matrix for multi-class classification
# Get class names
class_names = sorted(y_multi.unique())

# Create multi-class confusion matrix
multi_cm = confusion_matrix(y_test_multi, y_pred_multi)

# Plot multi-class confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(multi_cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Multi-Class Disease Classification')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Get feature importance for binary model
vectorizer = binary_pipeline.named_steps['vectorizer']
classifier = binary_pipeline.named_steps['classifier']

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Get feature importances
importances = classifier.feature_importances_

# Create a DataFrame for easier viewing
feature_importance = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
})

# Sort by importance
feature_importance = feature_importance.sort_values('importance', ascending=False)
print("\nTop 10 important features for respiratory disease prediction:")
print(feature_importance.head(10))
```
